# Capstone Project - The Battle of Neighborhoods (Week 1) #

***In this project, I am serching for the response for the following questions: 'Which appartment to choose?'
There is an italian man who wants to move to other country. He posses 2 appartments: one in Warsaw and second one in Paris.
He is fun of food and he hates cooking. Then, he wants to know which kind of food he can eat nearby each of these appartments.
He does not have a car so one of the important issues is the distance from flat to the restaurant.
For him, 1000metres is maximum distance to walk.
He likes trying new kind of food but of course as he is italian he loves italian kitchen.
Using foursquare I will try to help him in this decision.***
